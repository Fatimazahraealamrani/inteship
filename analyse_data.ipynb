```python
data_again=pd.read_csv('beauty_products.csv')
data_rating=pd.DataFrame({'Number of rating':data_again.groupby('ProductId').count()['Rating'],'Mean Rating':data_again.groupby('ProductId').mean()['Rating']})
data_rating.head(10)
```

```python
  #top 10 products based on sales.
popular_products=pd.DataFrame(data_again.groupby('ProductId')['Rating'].count())
most_pop=popular_products.sort_values('Rating',ascending=False)
most_pop.head(10)
```
